#### Exercice 1 : Prediction du prix d'une maison en fonction de sa surface

### 1. Importation des bibliotheques

```python
import pandas as pd # pour l'analyse de donneees
import matplotlib.pyplot as plt # pour la visualisation de donnnees
from sklearn import linear_model
from sklearn.metrics import root_mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore') 
```

### 2. Chargement des donnees

```python
df = pd.read_csv("prix_maisons.csv")
```

```python
plt.xlabel("Superficie")
plt.ylabel("Prix")

plt.scatter(df['Supervicie'], df['Prix'])
```

### 3. Creation du modele de regression lineaire

```python
lr = linear_model.LinearRegression()
```

```python
lr
```

#### 3.1 Entrainer le modele sur les donnees

```python
y = df['Prix']
X = df.drop(columns=['Prix'])
```

```python
lr.fit(X,y)
```

#### 3.2 Test de la prediction

```python
prix = lr.predict([[900]])
```

```python
prix
```

```python
predite = lr.predict(X)
```

```python
r2_score(y, predite)
```

```python
root_mean_squared_error(y,predite)
```

```python
plt.xlabel("Superficie")
plt.ylabel("Prix")

plt.scatter(df['Supervicie'], df['Prix'], color="red")
plt.plot(df['Supervicie'], lr.predict([[predite]]))
```
