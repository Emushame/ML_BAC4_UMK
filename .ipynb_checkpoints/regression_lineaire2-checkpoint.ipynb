## Prédire la note d’un étudiant selon le nombre d’heures d’étude

### 1. Importation des bibliotheques

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore') 
```

### 2. Importation de donnees

```python
df = pd.read_excel("notes_etudiant.xlsx")
```

```python
df
```

```python
df.columns
```

### 3. Separer les features et le label

```python
X = df[['Heures_etude', 'Participation_cours (%)', 'Devoirs_rendus']]
y = df['Note_finale (/100)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = LinearRegression()
model.fit(X_train, y_train)
```

### 4. Tester et predire 

```python
prediction = model.predict(X_test)
print("Prédictions :", prediction)
```

### 5. Evaluer le modele

```python
print("R2 Score:", r2_score(y_test, prediction))
print("MSE:", mean_squared_error(y_test, prediction))
```

### 6. Predire un nouvel eleve

```python
new_student = [[4.2, 72, 3]]
print("Note prédite :", model.predict(new_student))
```
